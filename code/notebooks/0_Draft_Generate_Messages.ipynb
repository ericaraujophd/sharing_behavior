```python
# Open JSON file with the information of the messages
json_string = "../data/messages/messages.json"
with open(json_string, 'r') as f:
    messages = json.load(f)
```

```python
# Load agents
json_string = "../data/agents/agents.json"
with open(json_string, 'r') as f:
    agents = json.load(f)
```

```python
# Generate 10 messages of each
qty_msgs = 10

messages_df = pd.DataFrame()
messages_df = pd.concat(
                [messages_df, pd.DataFrame([messages['fake_news']]*qty_msgs)], ignore_index=True)
messages_df = pd.concat(
                [messages_df, pd.DataFrame([messages['news']]*qty_msgs)], ignore_index=True)
messages_df = pd.concat(
                [messages_df, pd.DataFrame([messages['holidays_pics']]*qty_msgs)], ignore_index=True)
messages_df = pd.concat(
                [messages_df, pd.DataFrame([messages['online_course_ad']]*qty_msgs)], ignore_index=True)
messages_df = pd.concat(
                [messages_df, pd.DataFrame([messages['cats']]*qty_msgs)], ignore_index=True)

# Repeat the same sequence 5 times
messages_df = pd.concat([messages_df]*5, ignore_index=True)
```
