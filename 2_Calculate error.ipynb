```python
import model_v2 as model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
Run a sequence of messages for one agent with specific traits and an initial state
'''
def run_agent_model(message_seq=None, traits=None, states=None, title='0'):

	timesteps = 20
	delta_t = 1
	speed_factor = 0.5
	weightList=None
	alogistic_parameters=None
	

	# Initialize empty df
	inputsDF = pd.DataFrame()

	for message in message_seq:
		

		g, w, s = model.run_message(message=message, weightList=weightList, traits=traits, states=states, alogistic_parameters=alogistic_parameters, 
			speed_factor=speed_factor, delta_t = delta_t, timesteps = timesteps)

		states=[s['pp_cons'], s['pp_lib'], s['cs_cons'], s['cs_lib'], s['mood']]

		status_results = {}
		for node in g.nodes():
		    status_results[node] = g.node[node]['status']

		inputsDF = inputsDF.append(pd.DataFrame(status_results), ignore_index=True)

	#inputsDF[['pp_cons', 'pp_lib', 'mood', 'cs_cons', 'cs_lib', 'fs_s']].plot(figsize=((14,6)))
	inputsDF[['mood', 'fs_h', 'fs_s']].plot(figsize=((14,6)))
	#plt.show()
	plt.savefig('output'+title+'.png')
	return inputsDF
```
